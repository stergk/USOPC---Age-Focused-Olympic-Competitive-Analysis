```python
query = """
select class,
    competition_date,
    case when event_gender = 'Men' then 1 else 0 end as gender_male,
    case when event_gender = 'Women' then 1 else 0 end as gender_female,
    event_name,
    event_name_short,
    info_strada_sport_name as sport_name,
    case when medal in ('G', 'S', 'B') then 1 else 0 end as medal,
    noc_name as country_name,
    person_age_years as age,
    rank
from data
where rank is not null
    and person_age_years is not null
"""
```

```python
processed = sql.sqldf(query)
```

```python
processed.isnull().mean(0)
```

```python
country_stats = sql.sqldf("""
select country_name, event_name_short,
    avg(medal) as medal_rate,
    count(1) as participation_count
from processed 
where gender_male = 1 
group by 1,2
having medal_rate > 0.000
""")

country_medal_rate = country_stats[['country_name', 'event_name_short', 'medal_rate']].pivot(index='country_name', columns='event_name_short')
country_medal_rate.columns = country_medal_rate.columns.get_level_values(1)
country_participation_count = country_stats[['country_name', 'event_name_short', 'participation_count']].pivot(index='country_name', columns='event_name_short')
country_participation_count.columns = country_participation_count.columns.get_level_values(1)
```

```python
for table in [country_medal_rate, country_participation_count]:
    
    t = table.fillna(0).round(2)*100

    plt.figure(figsize=(15,15))
    sns.heatmap(data=t, annot=True, fmt='g', linewidths=.01)
```

```python
country_age_stats = sql.sqldf("""
select country_name, cast(age as integer) AS age, avg(medal) as medal_rate, count(1) as participation_count
from processed 
where gender_male = 1
group by 1,2
having avg(medal)>0.0
""")

country_age_medal_rate = country_age_stats[['country_name', 'age', 'medal_rate']].pivot(index='country_name', columns='age')
country_age_medal_rate.columns = country_age_medal_rate.columns.get_level_values(1)
country_age_participation_count = country_age_stats[['country_name', 'age', 'participation_count']].pivot(index='country_name', columns='age')
country_age_participation_count.columns = country_age_participation_count.columns.get_level_values(1)

for table in [country_age_medal_rate, country_age_participation_count]:
    
    t = table.fillna(0).round(2)*100

    plt.figure(figsize=(30,15))
    sns.heatmap(data=t, annot=True, fmt='g', linewidths=.01)
    
plt.figure(figsize=(30,15))
sns.heatmap(data=country_age_participation_count.divide(country_age_participation_count.sum(1), axis=0).fillna(0).round(3)*100, annot=True, fmt='g', linewidths=.01)

```

```python
query = """
select class, gender_male, gender_female, event_name_short, rank
from processed
"""
```

```python
to_test = sql.sqldf(query)
```

```python
to_test
```

```python
to_one_hot = ['class', 'event_name_short']
```

```python
to_test_one_hot = to_test.copy()
for hot in to_one_hot:
    unique = to_test_one_hot[hot].unique()[:-1]
    for label in unique:
        to_test_one_hot[label] = to_test_one_hot[hot].apply(lambda x: 1 if x==label else 0)
    to_test_one_hot.drop(hot, axis=1, inplace=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(to_test_one_hot.drop('rank', axis=1), to_test_one_hot['rank'], test_size=0.1, random_state=42)
```

```python
tree = DecisionTreeRegressor()
rf = RandomForestRegressor(n_estimators=1000)
```

```python
tree.fit(X_train, y_train)
rf.fit(X_train, y_train)
```

```python
preds = rf.predict(X_test)
preds_train = rf.predict(X_train)
```

```python
print(mean_absolute_error(y_test, preds))
print(mean_absolute_error(y_train, preds_train))
```

```python
preds = tree.predict(X_test)
preds_train = tree.predict(X_train)
```

```python
print(mean_absolute_error(y_test, preds))
print(mean_absolute_error(y_train, preds_train))
```
